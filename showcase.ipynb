```python
import phys_lab_calc as plc
import scipy.stats as sps #for noise generation
```

```python
U = list(plc.measurement(i + sps.norm.rvs(scale=5), 0.1 * i) for i in range(10, 101, 10))
I = list(plc.measurement(i, 0.1) for i in range(1, 11))
```

```python
R, _ = plc.dependency(I, U).plot(xlbl="I_on_resistor", xmu="A", ylbl="U_on_resistor", ymu="V", fitline=True)
```

```python
R.view(3)
```

```python
print("The resistance is", R, " Ohm")
```
